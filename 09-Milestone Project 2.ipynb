```python
import random

suits = ('Hearts', 'Diamond', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10,
          'King':10, 'Ace':11}
```

```python
class Card:
    
    def __init__(self, suit, rank):
      
      self.suit = suit
      self.rank = rank
      self.value = values[rank]
    
    def __str__(self):
      
      return self.rank + " of " + self.suit
    
```

```python
class Deck:
    
    def __init__(self):
      
      self.deck = []
      
      for suit in suits:
          for rank in ranks:
            
            created_card = Card(suit,rank)
            self.deck.append(created_card)
            
    def __str__(self):
      
      deck_show = ' '
      for card in self.deck:
        deck_show += card.__str__()
      return deck_show
            
    def shuffle(self):
      
      random.shuffle(self.deck)
      
    def deal_one(self):
      
      return self.deck.pop()
```

```python
class Hand:
    
    def __init__(self):
        
        self.cards = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
        
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1
        
    def adjust_for_ace(self):
        
        if self.value > 21 and self.aces >= 1:
            self.value -= 10
            self.aces -= 1
            
```

```python
class Chips:
    
    def __init__(self):
        self.amt = 100
        self.bet = 0
      
    def win_bet(self):
        self.amt += self.bet
        
    def lose_bet(self):
        self.amt -= self.bet
        
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input("Please provide a number: "))
        except:
            print("That is not a number")
            continue
        else:
            if(chips.bet > chips.amt):
                print("You cannot bet more chips than you have")
            else:
                break
            
```

```python
def hit(deck,hand):
    
    hand.add_card(deck.deal_one())
    hand.adjust_for_ace()
    
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    while True:
        check = input("Would you like to 'hit' or 'stand'?: ")
        
        if (check.lower() == 'hit'):
            
            hit(deck,hand)
        
        elif (check.lower() == 'stand'):
            
            print("Player Stands. Dealer is playing now.")
            playing = False
            
        else:
            
            print("Try again! Please only enter 'hit' or 'stand'!")
            continue
        
        break
            
```

```python
def show_some(dealer,player):
    
    print("\nDealer's Hand: ")
    print(" <hidden card>")
    print(f" {dealer.cards[1]}")
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print(f"Player's Hand = {player.value}")
    
def show_all(dealer,player):
    
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print(f"Dealer's Hand = {dealer.value}")
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print(f"Player's Hand = {player.value}")
    
```

```python
def player_blackjack(chips):
    print("Player has won by BLACKJACK")
    chips.win_bet()

def player_busts(chips):
    print("Player has BUST!")
    chips.lose_bet()

def player_wins(chips):
    print("Player has won the pot!")
    chips.win_bet()

def dealer_busts(chips):
    print("Dealer has BUST!")
    chips.win_bet()
    
def dealer_wins(chips):
    print("Dealer has won!")
    chips.lose_bet()
    
def push():
    print("Player and Dealer tie! PUSH!")
    
```

```python
 # Print an opening statement
print("Welcome to Blackjack!\nRules:\n 1. Get as close to 21 as possible without going over\n 2. Dealer hits upto 17\n 3. Aces count as 1 or 11 depending on the hand's favor\n 4. All Face Cards are worth 10")

while True:
    playing = True
    
    # Create & shuffle the deck, deal two cards to each player
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand()
    player_hand.add_card(deck.deal_one())
    player_hand.add_card(deck.deal_one())
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal_one())
    dealer_hand.add_card(deck.deal_one())
    
    player_hand.adjust_for_ace()
    
    # Set up the Player's chips
    chips = Chips()
    
    # Prompt the Player for their bet
    take_bet(chips)
    
    # Show cards (but keep one dealer card hidden)
    show_some(dealer_hand,player_hand)
    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(deck,player_hand)
        
        # Show cards (but keep one dealer card hidden)
        show_some(dealer_hand,player_hand)
 
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if player_hand.value > 21:
            
            player_busts(chips)
            show_all(dealer_hand,player_hand)
            break
    
        if player_hand.value == 21:
            
            player_blackjack(chips)
            show_all(dealer_hand,player_hand)
            break
            
    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
    if player_hand.value < 21:
        
        while(dealer_hand.value < 17):
            
            hit(deck,dealer_hand)
       
        # Show all cards
        show_all(dealer_hand,player_hand)
    
        # Run different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(chips)

        elif dealer_hand.value > player_hand.value:
            dealer_wins(chips)

        elif player_hand.value > dealer_hand.value :
            player_wins(chips)

        else:
            push()        
    
    # Inform Player of their chips total 
    print(f"Player has {chips.amt} chips left")
    
    # Ask to play again
    game_on = input("Enter 'y' if you would like to play another game: ")
    
    if game_on.lower() == 'y':
        continue
    
    else:
        
        print("Thanks for playing!")
        break
    
```
